```python
###word stemming###
from nltk.stem import PorterStemmer;
sentences=["fairly","beauty","history","going","goes","wenting","hunting","comes","entropy"]
ps=PorterStemmer()
for i in sentences:
    print(ps.stem(i))
```

```python
### regex stemmer####
from nltk.stem import RegexpStemmer
reg=RegexpStemmer('ing$|s$|able$|y$|es$')
sentences=["fairly","beauty","history","going","goes","wenting","hunting","comes","entropy"]
for i in sentences:
    print(reg.stem(i))
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
ps=PorterStemmer()
sb=SnowballStemmer('english')
sentences=["fairly","beauty","history","going","goes","wenting","hunting","comes","entropy"]
for i in sentences:
    print(sb.stem(i))
print(ps.stem("fairly"))
sb.stem('fairly')
```

```python
### parts of speech tagging###

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import pos_tag



# Sample text
text = "NLTK is a leading platform for building Python programs to work with human language data."

# Tokenize the text
tokens = word_tokenize(text)

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

# POS tagging on filtered tokens
pos_tags = pos_tag(filtered_tokens)

# Output results
# print("Original Tokens:", tokens)
# print("Filtered Tokens (No Stopwords):", filtered_tokens)
print("POS Tags:", pos_tags)
```

```python
# Example text
text = "Apple is looking at buying U.K. startup for $1 billion in 2025."

import nltk
from nltk import word_tokenize, pos_tag, ne_chunk

# Download required datasets
# nltk.download('punkt')
# nltk.download('maxent_ne_chunker')
# nltk.download('words')
# nltk.download('maxent_ne_chunker_tab')

text = "Barack Obama was born in Hawaii."

# Tokenize and tag
tokens = word_tokenize(text)
tags = pos_tag(tokens)

# Named Entity Recognition
tree = ne_chunk(tags)
print(tree)
```
